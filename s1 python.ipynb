s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])

> 교집합 s1 & s2 >> {4, 5, 6}

> 합집합 

s1.intersection(s2) >> {4, 5, 6}  s1 | s2  >> {1, 2, 3, 4, 5, 6, 7, 8, 9} s1.union(s2) = s2.union (s1)

> 차집합 -(빼기) , difference 함수를 사용한다. 

> 집합 함수 

- add 값 1개 추가 s1.add(4)
- update 값 여러 개 추가  s1.update([4, 5, 6])
- remove 특정값 제거 s1.remove(2)
- discard 특정 값 제거 s1.discard(2)
- clear  모든 값 제거 s1.clear()
